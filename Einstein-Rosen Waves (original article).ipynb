# Einstein-Rosen Waves

In this notebook we are going to compute the Einstein tensor associated to a **cylindrically symmetric, non stationary** metric which is Lorentzian in the entire manifold.
<!--  -->
The metric used in the **original article** from Albert Einstein and Nathan Rosen has the following form

$$\large ds^2 = A(\rho,t)(dt^2-d\rho^2)-B(\rho,t)d\phi ^2-C(\rho,t)dz^2$$


First, we define the manifold.

```python
#           (dim,name, latex name)
M = Manifold(4,'M',r'\mathcal{M}')
M
```

Now we define an open set on the manifold

```python
U = M.open_subset('U')
U
```

Now we define the homeomorphism $C: M \to \mathbb{R}^{4}$

```python
#chart_name.<coordinates> = open set.chart('coordinate:(range)')
C.<rh, z, ph, t> = U.chart(r'rh:(0,+oo):x_1 z:(-oo,+oo):x_2 ph:(0,2*pi):x_3  t:(0,+oo):x_4')
C
```

```python
M.default_chart()
```

Now we focus on declaring the metric

```python
g = M.lorentzian_metric('g')
```

We need to define two arbitrary functions $\gamma(\rho,t)$ and $\psi(\rho,t)$

```python
A = function('A',nargs = 2)
B = function('B',nargs = 2)
C = function('C',nargs = 2)
```

```python
g[0,0], g[1,1], g[2,2], g[3,3] = -A(rh,t),-B(rh,t),-C(rh,t),A(rh,t)

g.display()
```

### Einstein Tensor

```python
ric = g.ricci()
r = g.ricci_scalar()
G = ric-(1/2)*g*r
```

## New variables

We may rewrite the metric in terms of new variables defined by:

$$\large \begin{cases}
\alpha = \ln A \\
\beta = \frac{1}{2}\ln\left(\frac{B}{C}\right) \\
\gamma = \frac{1}{2}\ln(BC)
\end{cases}$$

So the metric takes the form

$$\large ds^2 = e^{\alpha}(dt^2-d\rho^2)-e^{\gamma-\beta}d\phi ^2-e^{\gamma+\beta}dz^2$$




```python
h = M.lorentzian_metric('h')
```

```python
a = function('a',nargs = 2)
b = function('b',nargs = 2)
c = function('c',nargs = 2)
```

```python
h[0,0], h[1,1], h[2,2], h[3,3] = -exp(a(rh,t)),-exp(c(rh,t)-b(rh,t)),-exp(c(rh,t)+b(rh,t)),exp(a(rh,t))
h.display()
```

```python
rich = h.ricci()
Rh = h.ricci_scalar()
Gh = rich-(1/2)*h*Rh
```

## Special coordinates


We are interested in transformations that do not spoil the conditions

$$\Large \begin{cases}
g_{14} = 0 \\
g_{11} = -g_{44}
\end{cases}$$

A necessarily and sufficent condition for this kind of transformations is 

$$\Large\begin{cases}
\frac{\partial \bar{x}_1}{\partial x_1} = \frac{\partial \bar{x}_4}{\partial x_4} \\
\frac{\partial \bar{x}_1}{\partial x_4} = \frac{\partial \bar{x}_4}{\partial x_1} 
\end{cases}$$

This implies that we may choose $\bar{x}_1 = (x_1,x_4)$ such that

$$\large \frac{\partial^2 \bar{x}_1}{\partial x_4}^2 - \frac{\partial^2 \bar{x}_1}{\partial x_1}^2 = 0$$

which is the wave equation for $\bar{x}_1$


Since $\sigma = e^\gamma$ follows the wave equation, there exists a transformation where

$$\Large \sigma = e^\gamma = k x_1$$

and since
$$\large \begin{matrix}
\gamma = \frac{1}{2}\ln B+\frac{1}{2}\ln C \\
\beta = \frac{1}{2}\ln B -\frac{1}{2}\ln C
\end{matrix}$$

we may identify $C = x_1^2$ and $B = k^2$, so the metric takes the form

$$\large ds^2 = e^{\alpha}(dx_4^2-dx_1^2)-k^2dx_2^2-x_1^2dx_3^2$$




```python
i = M.lorentzian_metric('i')
```

```python
i[0,0], i[1,1], i[2,2], i[3,3] = -exp(a(rh,t)),-exp(ln(rh)-b(rh,t)),-exp(ln(rh)+b(rh,t)),exp(a(rh,t))
i.display()
```

```python
rici = i.ricci()
Ri = i.ricci_scalar()
Gi = rici-(1/2)*i*Ri
```
