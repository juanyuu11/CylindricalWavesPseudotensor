# Einstein-Rosen Waves

In this notebook we are going to compute the Einstein tensor associated to a **cylindrically symmetric, non stationary** metric which is Lorentzian in the entire manifold.

A metric with the mentioned properties has the following form:

$$\large ds^2 = e^{2(\gamma-\psi)}(dt^2-d\rho^2)-\rho^2e^{-2\psi}d\phi ^2-e^{2\psi}dz^2$$

where $\gamma =  \gamma(\rho,t) $ and $\psi = \psi(\rho,t)$

First, we define the manifold.

```python
#           (dim,name, latex name)
M = Manifold(4,'M',r'\mathcal{M}')
M
```

Now we define an open set on the manifold

```python
U = M.open_subset('U')
U
```

Now we define the homeomorphism $C: M \to \mathbb{R}^{4}$

```python
#chart_name.<coordinates> = open set.chart('coordinate:(range)')
C.<t, rh, ph, z> = U.chart(r't:(0,+oo) rh:(0,+oo):\rho ph:(0,2*pi):\phi z:(-oo,+oo)')
C
```

```python
M.default_chart()
```

Now we focus on declaring the metric

```python
g = M.lorentzian_metric('g')
```

We need to define two arbitrary functions $\gamma(\rho,t)$ and $\psi(\rho,t)$

```python
# Funciones arbitrarias psi(t, rho), chi(t, rho), k(t, rho)
# Las pasamos como funciones de (t, x, y) via rho = sqrt(x^2+y^2)
psi_f = function('psi', nargs=2)
chi_f = function('chi', nargs=2)
#k_f   = function('k',   nargs=2)

PSI = psi_f(t, rh)
CHI = chi_f(t, rh)
```

```python
g[0,0], g[1,1], g[2,2], g[3,3] = exp(2*(CHI-PSI) ), -exp(2*(CHI-PSI ) ),-rh^2*exp(-2*PSI),-exp(2*PSI)

g.display()
```

```python
gI = g.inverse()
gI.display()
```

```python
nab = g.connection()
print(nab)
```
