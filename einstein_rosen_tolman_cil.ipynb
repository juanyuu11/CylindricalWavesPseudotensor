# Einstein-Rosen Waves — Coordenadas Cilíndricas
## Densidad Lagrangiana y Pseudotensor de Tolman

Partimos del elemento de línea en coordenadas cilíndricas $(t, \rho, \phi, z)$:

$$ds^2 = e^{2(\gamma-\psi)}\bigl(dt^2 - d\rho^2\bigr) - \rho^2 e^{-2\psi}\,d\phi^2 - e^{2\psi}\,dz^2$$

donde $\gamma = \gamma(\rho,t)$ y $\psi = \psi(\rho,t)$. Las componentes no nulas de la métrica son:

$$g_{tt} = e^{2(\gamma-\psi)}, \qquad g_{\rho\rho} = -e^{2(\gamma-\psi)}, \qquad g_{\phi\phi} = -\rho^2 e^{-2\psi}, \qquad g_{zz} = -e^{2\psi}$$

La **densidad lagrangiana** gravitacional es:
$$\mathcal{L} = \mathfrak{g}^{\mu\nu}\left(\Gamma^\sigma_{\rho\sigma}\Gamma^\rho_{\mu\nu} - \Gamma^\rho_{\mu\sigma}\Gamma^\sigma_{\nu\rho}\right)$$

y el **pseudotensor de Tolman**:
$$t^\nu{}_{\mu} = \frac{c^4}{16\pi G}\left(\mathfrak{g}^{\rho\lambda}{}_{,\mu}\,\frac{\partial\mathcal{L}}{\partial\,\mathfrak{g}^{\rho\lambda}{}_{,\nu}} - \delta^\nu_\mu\,\mathcal{L}\right)$$

## 1. Variedad y carta de coordenadas

```python
M = Manifold(4, 'M', r'\mathcal{M}')
U = M.open_subset('U')

C.<t, rh, ph, z> = U.chart(r't:(0,+oo) rh:(0,+oo):\rho ph:(0,2*pi):\phi z:(-oo,+oo)')
C
```

## 2. Funciones de la métrica

```python
# Funciones arbitrarias psi(t, rho), chi(t, rho), k(t, rho)
# Las pasamos como funciones de (t, x, y) via rho = sqrt(x^2+y^2)
psi_f = function('psi', nargs=2)
chi_f = function('chi', nargs=2)
#k_f   = function('k',   nargs=2)

PSI = psi_f(t, rh)
CHI = chi_f(t, rh)
```

## 3. Métrica de Einstein-Rosen

```python
g = M.lorentzian_metric('g')

g[0,0], g[1,1], g[2,2], g[3,3] = exp(2*(CHI-PSI) ), -exp(2*(CHI-PSI ) ),-rh^2*exp(-2*PSI),-exp(2*PSI)

g.display()
```

## 4. Métrica inversa $g^{\mu\nu}$

```python
gI = g.inverse()
gI.display()
```

## 5. Símbolos de Christoffel $\Gamma^\sigma_{\mu\nu}$

```python
nab = g.connection()
nab.display()
```

## 6. Determinante y densidad tensorial $\mathfrak{g}^{\mu\nu} = \sqrt{-g}\,g^{\mu\nu}$

```python
n = 4
coords   = [t, rh, ph, z]
cn       = ['t', 'rh', 'ph', 'z']   # nombres para imprimir

# Matriz de componentes
gLL = matrix([[g[i,j].expr() for j in range(n)] for i in range(n)])
gUU = [[gI[i,j].expr() for j in range(n)] for i in range(n)]

det_g      = gLL.det().simplify_full().factor()
sqrt_neg_g = sqrt(-det_g).simplify_full().factor()

print('det(g) =')
show(det_g)
print('sqrt(-g) =')
show(sqrt_neg_g)
```

```python
# frak_g^{munu} = sqrt(-g) * g^{munu}
frak_gUU = {}
for mu in range(n):
    for nu in range(mu, n):
        val = (sqrt_neg_g * gUU[mu][nu]).simplify_full()
        frak_gUU[(mu,nu)] = val
        frak_gUU[(nu,mu)] = val

print('Componentes no nulas de frak_g^{munu}:')
for mu in range(n):
    for nu in range(mu, n):
        val = frak_gUU[(mu,nu)]
        if val != 0:
            print(f'frak_g^{{{cn[mu]}{cn[nu]}}} =')
            show(val.factor())
```

## 7. Densidad Lagrangiana

$$\mathcal{L} = \mathfrak{g}^{\mu\nu}\left(\Gamma^\sigma_{\lambda\sigma}\Gamma^\lambda_{\mu\nu} - \Gamma^\lambda_{\mu\sigma}\Gamma^\sigma_{\nu\lambda}\right)$$

```python
# Extraer Christoffel del objeto conexion: nab[k,i,j] = Gamma^k_{ij}
Gamma = [[[nab[k,i,j].expr() for j in range(n)] for i in range(n)] for k in range(n)]
```

```python
# Término 1: g^{munu} Gamma^s_{lam s} Gamma^{lam}_{munu}
term1 = SR(0)
for mu in range(n):
    for nu in range(n):
        if gUU[mu][nu] == 0: continue
        for lam in range(n):
            for s in range(n):
                term1 += gUU[mu][nu] * Gamma[s][lam][s] * Gamma[lam][mu][nu]

# Término 2: g^{munu} Gamma^{lam}_{mu s} Gamma^s_{nu lam}
term2 = SR(0)
for mu in range(n):
    for nu in range(n):
        if gUU[mu][nu] == 0: continue
        for lam in range(n):
            for s in range(n):
                term2 += gUU[mu][nu] * Gamma[lam][mu][s] * Gamma[s][nu][lam]

L = (term1 - term2).simplify_full()

print('Densidad Lagrangiana L =')
show(L)
```

## 8. Pseudotensor de Tolman

Necesitamos $\partial\mathcal{L}/\partial(\mathfrak{g}^{\rho\lambda}{}_{,\nu})$. Para ello reexpresamos $\mathcal{L}$ usando variables auxiliares $\delta g_{\mu\nu,\alpha}$ para las derivadas de la métrica covariante, diferenciamos algebraicamente, y luego sustituimos los valores reales.

```python
# ─────────────────────────────────────────────────────────────────
# Sección 8 — Pseudotensor de Tolman
# usando las derivadas analíticas del Lagrangiano
#
#   ∂L/∂(g^{μν})       = Γ^λ_{μρ} Γ^ρ_{νλ}  −  Γ^ρ_{μν} Γ^λ_{ρλ}
#
#   ∂L/∂(g^{μν}_{,ρ})  = Γ^ρ_{μν}
#                       − ½ δ^ρ_μ Γ^λ_{νλ}
#                       − ½ δ^ρ_ν Γ^λ_{μλ}
# ─────────────────────────────────────────────────────────────────

# ── Derivadas ∂L/∂(frak_g^{μν}) ──────────────────────────────────
dL_dfrak = {}
for mu in range(n):
    for nu in range(mu, n):
        val = SR(0)
        for lam in range(n):
            for rho in range(n):
                # Γ^λ_{μρ} Γ^ρ_{νλ}
                val += Gamma[lam][mu][rho] * Gamma[rho][nu][lam]
                # − Γ^ρ_{μν} Γ^λ_{ρλ}
                val -= Gamma[rho][mu][nu] * Gamma[lam][rho][lam]
        val = val.simplify_full()
        dL_dfrak[(mu, nu)] = val
        dL_dfrak[(nu, mu)] = val   # simetría

# ── Derivadas ∂L/∂(frak_g^{μν}_{,ρ}) ────────────────────────────
dL_ddfrak = {}
for mu in range(n):
    for nu in range(mu, n):
        for rho in range(n):
            val = SR(0)
            # Γ^ρ_{μν}
            val += Gamma[rho][mu][nu]
            # − ½ δ^ρ_μ Γ^λ_{νλ}
            if rho == mu:
                val -= SR(1)/2 * sum(Gamma[lam][nu][lam] for lam in range(n))
            # − ½ δ^ρ_ν Γ^λ_{μλ}
            if rho == nu:
                val -= SR(1)/2 * sum(Gamma[lam][mu][lam] for lam in range(n))
            val = val.simplify_full()
            dL_ddfrak[(mu, nu, rho)] = val
            dL_ddfrak[(nu, mu, rho)] = val   # simetría en μν

print('Derivadas del Lagrangiano calculadas.')
```

```python
# ── Derivadas de frak_g^{ρλ} respecto a coordenadas ─────────────
dfrak = {}
for r in range(n):
    for l in range(r, n):
        val_rl = frak_gUU[(r, l)]
        for mu in range(n):
            d = diff(val_rl, coords[mu]).simplify_full()
            dfrak[(r, l, mu)] = d
            dfrak[(l, r, mu)] = d

# ── Pseudotensor de Tolman ────────────────────────────────────────
#
#   t^ν_μ = (c⁴/16πG) [ frak_g^{ρλ}_{,μ} · ∂L/∂(frak_g^{ρλ}_{,ν})
#                      − δ^ν_μ · L ]
#
c_light, G_N = var('c G_N', domain='positive')
prefactor = c_light^4 / (16 * pi * G_N)

t_Tolman = [[SR(0)]*n for _ in range(n)]

for nu in range(n):
    for mu in range(n):
        acum = SR(0)
        for r in range(n):
            for l in range(r, n):
                factor = 1 if r == l else 2   # simetría de frak_g^{rl}
                acum += factor * dfrak[(r, l, mu)] * dL_ddfrak[(r, l, nu)]
        delta = L if nu == mu else SR(0)
        t_Tolman[nu][mu] = (prefactor * (acum - delta)).simplify_full()

print('Pseudotensor de Tolman calculado.')
```

## 9. Resultados

```python
print('Densidad Lagrangiana L =')
show(L.factor())
```

```python
for nu in range(n):
    for mu in range(n):
        val = t_Tolman[nu][mu]
        lbl = f't^{{{cn[nu]}}}_{{{cn[mu]}}}'
        if val == 0:
            print(lbl, '= 0')
        else:
            print(lbl, '=')
            show(val.factor())
```

## 10. Verificaciones

```python
# Traza
traza = sum(t_Tolman[nu][nu] for nu in range(n)).simplify_full()
print('Traza t^mu_mu =')
show(traza)

# Densidad de energia gravitacional
print('t^t_t =')
show(t_Tolman[0][0].factor())

# Presion en z
print('t^z_z =')
show(t_Tolman[3][3].factor())
```
